**Questão 1 — Classificação de Temperaturas (Condicionais + Loop)
(0,5 ponto)**

Dada a lista de temperaturas registradas durante a semana:

temperaturas = [18, 22, 30, 35, 28, 15, 40]

Classifique cada dia como:

•	Abaixo de 20 → "Frio"

•	Entre 20 e 30 → "Agradável"

•	Acima de 30 → "Quente"

Tarefas:

1.	Criar uma lista com as classificações
2.	Contar quantos dias houve em cada categoria


```python
temperaturas = [18, 22, 30, 35, 28, 15, 40]
classificacoes = []
contagem_categorias = {"Frio": 0, "Agradável": 0, "Quente": 0}
for temp in temperaturas:
    if temp < 20:
        classificacao = "Frio"
    elif 20 <= temp <= 30:
        classificacao = "Agradável"
    else:
        classificacao = "Quente"
    classificacoes.append(classificacao)
    contagem_categorias[classificacao] += 1

print("Classificações das temperaturas:", classificacoes)
print("Contagem por categoria:", contagem_categorias)
```

**Questão 2 — Sistema de Avaliação de Alunos
(0,5 ponto)**

Dada a lista de notas finais:

notas = [4.5, 6.0, 7.8, 9.0, 5.5, 8.2]

Classifique:

•	< 5 → "Reprovado"

•	5 a 7 → "Recuperação"

•	7 → "Aprovado"

Tarefas:

1.	Gerar lista de classificações
2.	Calcular:
o	Quantidade de aprovados
o	Percentual de reprovados


```python
notas = [4.5, 6.0, 7.8, 9.0, 5.5, 8.2]

reprovados_count = 0
recuperacao_count = 0
aprovados_count = 0

for n in notas:
  if n < 5:
    print(f"Aluno Reprovado! Nota {n}.")
    reprovados_count += 1
  elif 5 <= n <= 7:
    print(f"Aluno em Recuperação! Nota {n}")
    recuperacao_count += 1
  else:
    print(f"Aluno Aprovado! Nota {n}")
    aprovados_count += 1

total_alunos = len(notas)

print(f"Quantidade de alunos aprovados: {aprovados_count}")

percentual_reprovados = (reprovados_count / total_alunos) * 100
print(f"Percentual de alunos reprovados: {percentual_reprovados:.2f}%")
```

**Questão 3 — Monitoramento de Consumo de Energia
(0,5 ponto)**

Consumo diário (kWh):

consumo = [120, 95, 200, 180, 75, 220, 160]

Regras:

•	< 100 → "Baixo"

•	100–180 → "Moderado"

•	180 → "Alto"

Tarefas:

1.	Filtrar dias com consumo alto
2.	Calcular:
o	Total de consumo da semana
o	Média semanal
3.	Exibir alerta se houver mais de 2 dias com consumo alto


```python
consumo = [120, 95, 200, 180, 75, 220, 160]
consumo_total = 0
quantidade_consumo_alto = 0
for c in consumo:
  consumo_total += c
  if(c < 100):
    print(f"Consumo {c}: Baixo")
  elif(100 <= c <= 180):
    print(f"Consumo {c}: Moderado")
  else:
    print(f"Consumo {c}: Alto")
    quantidade_consumo_alto = quantidade_consumo_alto + 1

print(f"Consumo total da semana: {consumo_total}")
quantidade = len(consumo)
média = consumo_total / quantidade
print(f"Consumo médio da semana: {média:.2f}")

if (quantidade_consumo_alto > 2):
  print("Alerta! Muitos dias com consumo alto!")
else:
  print("Semana normal.")
```

**Questão 4 — Simulação de Carrinho de Compras
(0,5 ponto)**

Lista de preços:

precos = [50, 120, 30, 200, 80, 15]

Regras de desconto:

•	Item < 50 → 5% desconto

•	50–150 → 10% desconto

•	150 → 15% desconto

Tarefas:

1.	Criar lista com preços finais
2.	Calcular:
o	Total da compra com desconto
o	Economia total obtida


```python
precos = [50, 120, 30, 200, 80, 15]
preço_total = 0
precos_finais =[]
for p in precos:
  preço_total += p
  if(p < 50):
    print(f"5% desconto")
    valor = p * 0.95
    print(f"Valor com desconto: {valor}")
  elif(50 <= p <= 150):
    print(f"10% desconto")
    valor = p * 0.9
    print(f"Valor com desconto: {valor}")
  else:
    print(f"15% desconto")
    valor = p * 0.85
    print(f"Valor com desconto: {valor}")
  precos_finais.append(valor)

print(f"Lista de preços com descontos da semana: {precos_finais}")
print(f"Preço total da semana: {sum(precos_finais)}")
quantidade = len(precos)
```
